```python
def line_left(X,Y,x,y): 
    k = (Y[0] - Y[1]) / (X[0]-X[1])
    b = Y[1] - k * X[1]
    if y > k * x + b:
        return True
    else:
        return False
```

```python
X_red = X[y==0]
X_yellow = X[y==1]
```

```python
plt.plot([-1.5,0.5],[-1.0,1.5])
plt.scatter(0.25,0)
print(line_left([-1.5,0.5],[-1.0,1.5],-1.5,0))
```

```python
mask = np.array([line_left([-1.5, 0.5], [-1.0, 1.5], x[0],x[1]) for x in X_red])
X_red_1 = X_red[mask]
```

```python
plt.scatter(X_red_1[:,0], X_red_1[:,1],c='red')
```

```python
mask = np.array([not line_left([-1.5, 0.5], [-1.0, 1.5], x[0],x[1]) for x in X_red])
X_red_2 = X_red[mask]
```

```python
plt.scatter(X_red_2[:,0], X_red_2[:,1],c='red')
```

```python
mask = np.array([line_left([1.0,2.5],[-0.25,1.5], x[0],x[1]) for x in X_yellow])
X_yellow_1 = X_yellow[mask]
```
